```python
df_customer=pd.DataFrame({
    'id':[1,2,3,4],
    'name':["Tom","Jerry","james","Dan"]
    
    
})
```

```python
df_info_2 = pd.DataFrame({
    'customer_id': [2,3,4,5],
    'age': [31,20,40,70],
    'sex': ['F', 'M', 'M', 'F']
})

pd.merge(
  df_customer, #we didn't have same coloms so we are saying what should take from left and right
  df_info_2, 
  left_on='id',   #left tabilil eth colom veneam
  right_on='customer_id'   #right tabilil eth colom venam
)
```

```python
#left join
```

```python
df_info_2 = pd.DataFrame({
    'id': [2,3,4,5],
    'age': [31,20,40,70],
    'sex': ['F', 'M', 'M', 'F']
})
```

```python
pd.merge(df_customer,df_info_2,how='left',on='id')   #how koduthillenkill default aayit inner aayirikkum
```

```python
pd.merge(df_customer,df_info_2,how='right',on='id')
```

```python
pd.merge(df_customer,df_info_2,how='outer',on='id')
```
