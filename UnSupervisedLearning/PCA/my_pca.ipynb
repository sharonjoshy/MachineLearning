```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data=pd.read_csv("HR.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True,ignore_index=True)
```

```python
data.shape
```

```python
# sales is an object column

data["sales"].nunique()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.corr()
```

```python
X = data.drop(["left"],axis=1).values   #values mathram edukkunnath numpy array aakan because speed kudan
y=data["left"].values
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

pca =PCA(n_components=2)
X_pca = pca.fit_transform(X)  


print("The original shape :",X.shape)
print("The new PCA shape :",X_pca.shape)
X_train,X_test,y_train,y_test = train_test_split(X_pca,y,test_size=0.3,random_state=10)

ss = SVC()
ss.fit(X_train,y_train)

print("Train score :",ss.score(X_train,y_train))
print("Test score :",ss.score(X_test,y_test))
```

```python
# pca=3

pca =PCA(n_components=3)
X_pca = pca.fit_transform(X)  


print("The original shape :",X.shape)
print("The new PCA shape :",X_pca.shape)
X_train,X_test,y_train,y_test = train_test_split(X_pca,y,test_size=0.3,random_state=10)

ss = SVC()
ss.fit(X_train,y_train)

print("Train score :",ss.score(X_train,y_train))
print("Test score :",ss.score(X_test,y_test))


```
