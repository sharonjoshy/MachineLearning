```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days,max_t, label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc="best")
plt.grid()
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')    #heading of the graph

plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc="best")   #it will automatically move loc koduthillenkilum
plt.grid()   # square box varum accurate value ariyan
plt.show()
```
