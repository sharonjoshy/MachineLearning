```python

# Loading Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

from sklearn.model_selection import GridSearchCV
```

```python
pd.read_
```

```python
data=pd.read_csv(r"C:\Users\sharo\Desktop\AI\AI Library\SupervisedLearning\9_Logistic_Regression_multiclass\breast_cancer_data.csv")
data
```

```python
data.drop(["id",'Unnamed: 32'],axis=1,inplace=True)

# using map function
data["diagnosis"]=data["diagnosis"].map({"B":0,"M":1})

X = data.drop("diagnosis",axis=1)
y=data["diagnosis"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
log = LogisticRegression()
rnd = RandomForestClassifier(n_estimators=500)
adaboost = AdaBoostClassifier()
gboost = GradientBoostingClassifier()
svm = SVC()
```

```python
# Randomforest 

forest_hypparams=[{'n_estimators':[50,80,100,150,200,300],'criterion':['gini', 'entropy', 'log_loss'],'max_depth':list(range(10,15)),'max_features':list(range(5,20))}]

#max_features  means ethra column venam in one tree,should be less than no of main columns of train data
#
```

```python
grid_clf = GridSearchCV(rnd,forest_hypparams,cv=10,scoring="accuracy")   #cv ---kfold
grid_clf.fit(X_train,y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.best_score_
```

```python

for clf in (log,rnd,svm,adaboost,gboost):
    clf.fit(X_train,y_train)

    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__,accuracy_score(y_test,y_pred))

    
    
    print("************************")
```
