```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
```

```python
boston = load_boston()
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df["Price"]=boston.target
boston_df.head()
```

```python
boston.keys()
```

```python
boston_df.isnull().sum()
```

```python
X=boston_df.drop(["Price"],axis=1)
y=boston_df["Price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_reg = DecisionTreeRegressor()   # data overfit aan 

dt_reg.fit(X_train,y_train)

sc=dt_reg.score(X_train,y_train)
print("R-squared Train",sc)

ss=dt_reg.score(X_test,y_test)
print("R-squared Test",ss)

y_pred = dt_reg.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE:",mse)
print("RMSE :",mse**(1/2.0))
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))

tree.plot_tree(dt_reg,filled=True,class_names=True)
plt.show()
```
